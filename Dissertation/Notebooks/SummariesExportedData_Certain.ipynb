```python
fig,(ax,bx,cx) = plt.subplots(3,1,figsize=(7,9),
                                 gridspec_kw = {'wspace':0, 'hspace':0})

caEpisodes = []
pEpisodes = []
colors = ['b','r','g','m']
labels_3 = ['Agent 1','Agent 2','Agent 3','Agent 4']
labels_1 = []
labels_2 = []
style = ['-','-','-','-']
for file,c,s in zip(tsFiles,colors,style):

    data = pd.read_csv(path_ts+"/"+file,index_col=0)
    data.columns = np.array(data.columns).astype(float).astype(int)
    # Alternative way of evaluating the effcientcy of dealing with crowded areas
    
    test = data.loc[3][data.loc[3]>=3]
    test_2 = data.loc[3][data.loc[3]>=6]
#     print(test.index[-1])
    test_1 = data.loc[2][data.loc[3]>=3]
    print("Average number of episodes with crowded areas %i" % ((test.index[-1]-test.index[0])/test_1.mean()))
    caEpisodes.append((test.index[-1]-test.index[0])/test_1.mean())
    pEpisodes.append((test_2.index[-1]-test_2.index[0])/test_1.mean())
    labels_1.append("My: %.2f,\nSigma: %.2f" % (data.iloc[0,49:].mean(),data.iloc[0,49:].std()))
    labels_2.append("My: %.2f,\nSigma: %.2f" % (data.iloc[1,49:].mean(),data.iloc[1,49:].std()))
    
    ael = []
    acr = []
    ma = 3
    maRange = np.arange(ma,len(data.loc[1]),ma)
    step = 0
    
    for obs in maRange:
        
        acr.append(np.array(data.iloc[0,step:obs]).mean())
        ael.append(np.array(data.iloc[1,step:obs]).mean())
        
        step += ma
    
    ax.plot(acr,c = c,linestyle=s)
    bx.plot(ael,c = c,linestyle=s)
    cx.plot(data.loc[3],c = c,linestyle=s)

# Add labels
ax.legend(labels_1,fontsize=12, bbox_to_anchor = (1,1))
bx.legend(labels_2,fontsize=12, bbox_to_anchor = (1,1))
cx.legend(labels_3,fontsize=12, bbox_to_anchor = (1,1))

# Set y limit
ax.set_ylim((-0.8,1))
bx.set_ylim(0,170)
cx.set_ylim(-0.1,7.8)

# Set axis labels
ax.set_ylabel('Average Cumulative\nReward',fontsize=14)
bx.set_ylabel('Average Episode\nLength',fontsize=14)
cx.set_ylabel('Lessons',fontsize=14)
cx.set_xlabel('Steps',fontsize=14)
# plt.savefig('graphs/Certain_comparison_100.jpg',bbox_inches='tight',dpi=100)
plt.show()
```

```python
# collisions = []
# episodes = []
ma = 500
labels_2 = []
collisionHolder = []
episodesHolder = []
pedestrianEpiHolder = []
pedestriansHolder = []

colors = ['b','r','g','m']
labels_1 = ['Agent 1','Agent 2','Agent 3','Agent 4']
labels_2 = []

fig,ax = plt.subplots(1,1,figsize=(8,4),gridspec_kw = {'wspace':0, 'hspace':0.1})

for i,FoI in enumerate([28,29,30,31]):#6,
    
    collisions = []
    pedestrians = []
    pedeEpisodes = []
    episodes = []
    maSeriesHolder = []
    path = "../../Exported_Data/"+availableFolders[FoI]

    files = os.listdir(path)
    files = [file for file in files if any([True if neededfile in file else False \
                                            for neededfile in ['stepsCA.txt',
                                                               'stepsToGoal.txt',
                                                               'pedCount.txt']])]
    for filesOfInterest in ['stepsCA.txt','stepsToGoal.txt','pedCount.txt']:

        processing = [file for file in files if filesOfInterest in file]

        for file in processing:
            with open(path+"/"+file,"r") as f:
                content = f.readlines()
            content = [re.split(",",re.sub("[ \n]","",obs)) for obs in content]

            if (len(content[0])) == 1:
                v,c = np.unique(np.array(content),return_counts = True)
                pedestrians.append(c[1])
                pedeEpisodes.append(sum(c))
            elif len(content[0]) == 2:

                collisions.append(len(content))

            elif len(content[0]) == 3:
                maSeries = []
                step = 0
                episodes.append(len(content))

                first = []
                second = []
                third = []

                for obs in content:

                    first.append(obs[0])
                    second.append(obs[1])
                    third.append(obs[2])

                first = np.array(first).astype(int)
                second = np.array(second).astype(int)
                third = np.array(third).astype(int)
    
                maRange = np.arange(ma,len(first))

                for maObs in maRange:

                    maFirst = first[step:maObs]
                    maThird = third[step:maObs]

                    s,f = np.unique(maThird,return_counts = True)
                    daGoal = sum([True if ((j == 1) and (i == 1)) else False for i,j in \
                                  zip(maFirst,maThird)])
    
                    if len(f) == 1:
    
                        maSeries.append(s)
    
                    else:
    
                        maSeries.append(daGoal/f[1])
    
                    step += 1
                maSeriesHolder.append(maSeries)
    
        if episodes != []:
            avgMaSeries = []    
            for obs in np.arange(max(episodes)-ma):
                temp = []
    
                # First series
                try:
                    temp.append(maSeriesHolder[0][obs])
                except:
                    None
                # Second series
                try:
                    temp.append(maSeriesHolder[1][obs])
                except:
                    None
    
                # Third series
                try:
                    temp.append(maSeriesHolder[2][obs])
                except:
                    None
    
                # Fourth series
                try:
                    temp.append(maSeriesHolder[3][obs])
                except:
                    None
    
                avgMaSeries.append(np.array(temp).mean())
    
    labels_2.append("Agent %i:\nMy: %.2f,\nSigma: %.2f" % (i+1,np.array(avgMaSeries)[2000:].mean(),
                                            np.array(avgMaSeries)[2000:].std()))
    ax.plot(avgMaSeries,c=colors[i])
    episodesHolder.append(episodes)
    collisionHolder.append(collisions)
    pedestriansHolder.append(pedestrians)
    pedestrianEpiHolder.append(pedeEpisodes)
ax.set_xlabel('Increments, Each 500 steps',fontsize=14)
ax.set_ylabel('Efficiency On\nDifficult Areas',fontsize=14)
ax.legend(np.unique(np.array(labels_2)),loc=1,fontsize=12,bbox_to_anchor=(1.28,1))
# plt.savefig('graphs/Certain_comparison_custom_100.jpg',bbox_inches='tight',dpi=100)
plt.show()
```
